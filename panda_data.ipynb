```python
import pandas as pd
from pandas_profiling import ProfileReport


df = pd.read_csv('gx_new .csv')

report = ProfileReport(df, title='Your Data Profiling Report', explorative=True)

# Convert the report to JSON format
report_json = report.to_json()

# Save the JSON report to a file
with open('your_data_profile.json', 'w') as json_file:
    json_file.write(report_json)

# Display HTML report
report_html = report.to_html()
print(report_html)
```
